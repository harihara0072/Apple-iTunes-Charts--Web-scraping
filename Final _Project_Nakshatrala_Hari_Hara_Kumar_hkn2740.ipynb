### Algorithm:

1. Creating a list of all the catogories to iterate and scrape the data.
2. Open a file in write mode so that if the file already exists then it will overwrite the old file
3. Iterate through each catagory and using the Beautiful Soup to first finding the div , and then the un-ordered list HTML tag that contains the data.
4. Seperate the three coloums that we need and adding it to the dictionary usig index as the key and the remaining two coloumns as the values.
5. Finally asking the user to enter the option he/she wants and using that as the key of the dictionary retreiving the data.

```python
#Importing the required libraries
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup

#Using the default part of the url to append to the catagory later
url = 'https://www.apple.com/itunes/charts/'


#Saving all the topics to iterate later
topics = ['songs/', 'albums/', 'free-apps/', 'paid-apps/', 'top-grossing-apps/', 'paid-books/', 'movies/', 'tv-shows/', 'music-videos/']


#Dictionary to save the reselt for each catagory
total_dict = {}

#file open in write mode to save the data. This will also overwrite if it already exists
file_name = open('rankings.txt', 'w', encoding='utf-8')


#For each topic
for topic in topics:
    
    #Writing the file name in the file
    file_name.write(topic.strip('/') + "\n")
    
    #Opening the HTML content 
    html = urllib.request.urlopen(url + topic)
    soup = BeautifulSoup(html, 'html.parser')
    
    #Finding the div, ul and li tags content using Beautiful soup
    soup_obj_div = soup.body.findAll('div',{'class':'section-content'})[1]
    soup_obj_ul = soup_obj_div.findAll('ul')[0]
    soup_obj_li = soup_obj_ul.findAll('li')
    
    
    #Dictionary for each topic
    new_dict = {}
    
    
    #Using the for loop to get 100 data
    for li in soup_obj_li:
        
        #Geting the index
        index = li.contents[0].contents[0].strip('.')
        
        #Geting the second coloum
        name = li.contents[2].find('a').contents[0]
        
        #Geting the third coloum
        album = li.contents[3].find('a').contents[0]
        
        #Adding that in a dictionary
        new_dict[index] = [name, album]
        
        #Witing data to the file
        file_name.write(index + ", " + name + ", " + album + "\n")
        
        
    #Adding the topic dictionary to the overall dictionary
    total_dict[topic.strip('/')] = new_dict

#Closing the file
file_name.close()

#Printing the content is saved in the file
print("The data is stored in file rankings.txt")

#Using the infinite while loop to give the options to the user
try:
    while True:
        
        #Asking the user to choose an option
        case = input("What is the category that you are interested in getting data \n (1) Songs \n (2) Albums \n (3) Free Applications \n (4) Paid Applications \n (5) Top Grossing Applications \n (6) Books \n (7) Movies \n (8) TV Shows \n (9) Music Videos \n (10) Exit")
        
        #Reading the user input and using the option as a key to dictonary
        if int(case) == 1:
            key = "songs"
        elif int(case) == 2:
            key = "albums"
        elif int(case) == 3:
            key = "free-apps"
        elif int(case) == 4:
            key = "paid-apps"
        elif int(case) == 5:
            key = "top-grossing-apps"
        elif int(case) == 6:
            key = "paid-books"
        elif int(case) == 7:
            key = "movies"
        elif int(case) == 8:
            key = "tv-shows"
        elif int(case) == 9:
            key = "music-videos"
        elif int(case) == 10:
            print("Good Bye!")
            break
            
        #If the interger is in range
        if(0 < int(case) < 11):
            
            #Asking the rank the user wants
            key2 = input("What ranking are you interested in getting data about:")
            
            #Checking if the value in range
            if (0 < int(key2) < 101):
                print("Information about ranking " + key2 + ": ", total_dict[key][key2][0] + ", " +total_dict[key][key2][1], '\n')
            else:
                
                #If not in range
                print("Invalid choice \n")
        #If not in range
        else:
            print("Invalid choice \n")
            
#If not an integer at all
except:
    print("Invalid choice \n")
    
        
    
        
        
```
